```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from astropy.time import Time
```

```python
import pandas as pd
redshift_df = pd.read_csv("redshifts.csv")
vou_df = pd.read_csv("vou_100.csv")

df = pd.merge(vou_df, redshift_df, how='left', left_on='source_name', right_on='name')
df
```

```python
df.drop(columns=['name', 'Unnamed: 0', 'source_name', 'Catalog'], inplace=True)
```

```python
df
```

```python
df['start_time']
```

```python
df['time'] = Time(df['end_time'], format='mjd').iso
df['time'] = pd.to_datetime(df['time'])
df
```

```python
df['year'] =df['time'].dt.year
df
```

```python
df.drop(columns=['start_time', 'end_time', 'time'], inplace=True)
df
```

```python
df['flag'].unique()
```

```python
flag_mapping = {'Det': 1, 'UL': 0}
df['flag_numeric'] = df['flag'].map(flag_mapping)
df.drop(columns=['flag'], inplace=True)
df
```

```python
numerical_variables = df.select_dtypes(include='number').columns
sns.heatmap(df[numerical_variables].corr(), annot = True)
```

```python
df.isna().sum()
```

```python
columns_to_plot = ['frequency', 'nufnu', 'nufnu_upper', 'nufnu_lower', 'flag_numeric', 'nufnu_err', 'redshift', 'year']

fig, axes = plt.subplots(nrows=4, ncols=2, figsize=(12, 16))
axes = axes.flatten()
sample_size = 1000
sampled_df = df.sample(n=sample_size, random_state=42)

for i, column in enumerate(columns_to_plot):
    column_data = sampled_df[column]
    sns.histplot(column_data, ax=axes[i], kde=True, bins=50)
    axes[i].set_title(column)

plt.tight_layout()
plt.show()
```
